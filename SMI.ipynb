```python
import numpy as np
```

```python
import pandas as pd
```

```python
data_pa = pd.read_csv("SMImonthly.csv",header=None, delimiter=';')
```

```python
data_np = data_pa.values
```

```python
monthly_returns = data_np[:, -1]
```

```python
median = np.median(monthly_returns)
```

```python
first_quartile = np.percentile(monthly_returns, 25)
```

```python
arithmetic_mean = np.mean(monthly_returns)
```

```python
starting_value = data_np[0, 1]
starting_value
```

```python
terminating_value = data_np[-1, 0]
terminating_value
```

```python
geometric_mean = (terminating_value*starting_value) ** (1/len(monthly_returns))-1
geometric_mean
```

```python
standard_deviation = np.std(monthly_returns)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd



datums_index = pd.date_range(start='1/1/2014', periods=len(data_np[:,1]), freq='M')

plt.figure(figsize=(10, 6))
plt.plot(datums_index, data_np[:,1], label='Monatliche Kurse')
plt.title('Monatliche Kurse über die Jahre')
plt.xlabel('Jahr')
plt.ylabel('Kurs')
plt.legend()
plt.grid(True)

# Zeigen Sie das Diagramm an
plt.show()
```

```python
datums_index = pd.date_range(start='1/1/2014', periods=len(monthly_returns), freq='M')

plt.figure(figsize=(10, 6))
plt.plot(datums_index, monthly_returns, label='Monatliche Renditen')
plt.title('Monatliche Renidten über die Jahre')
plt.xlabel('Jahr')
plt.ylabel('Kurs')
plt.legend()
plt.grid(True)
plt.axhline(y=median, color='r', linestyle='-', label='Median')
plt.axhline(y=arithmetic_mean, color='g', linestyle='-', label='Arithmetisches Mittel')
plt.axhline(y=geometric_mean, color='y', linestyle='-', label='Geometrisches Mittel')
plt.legend()
# Zeigen Sie das Diagramm an
plt.show()
```

```python
forecast_return = np.random.normal(arithmetic_mean, standard_deviation, 120)
forecast_return = np.around(forecast_return, decimals=2)
forecast_return
```

```python
combined_returns = np.concatenate((monthly_returns, forecast_return))
```

```python
datums_index = pd.date_range(start='1/1/2014', periods=len(combined_returns), freq='M')

plt.figure(figsize=(10, 6))
plt.plot(datums_index, combined_returns, label='Monatliche Renditen')
plt.title('Monatliche Renidten über die Jahre')
plt.xlabel('Jahr')
plt.ylabel('Kurs')
plt.legend()
plt.grid(True)
plt.axhline(y=median, color='r', linestyle='-', label='Median')
#plt.axhline(y=arithmetic_mean, color='g', linestyle='-', label='Arithmetisches Mittel')
#plt.axhline(y=geometric_mean, color='y', linestyle='-', label='Geometrisches Mittel')
plt.legend()

plt.show()
```

```python
monthly_values = data_np[:,1]
```

```python
actual_changes = (monthly_values[-1] * (1+np.cumsum(forecast_return/100)))
```

```python
combined_values = np.concatenate((monthly_values, actual_changes))
```

```python
datums_index = pd.date_range(start='1/1/2014', periods=len(combined_values), freq='M')

plt.figure(figsize=(10, 6))
plt.plot(datums_index, combined_values, label='Monatliche Kurse')
plt.title('Monatliche Kurse über die Jahre')
plt.xlabel('Jahr')
plt.ylabel('Kurs')
plt.legend()
plt.grid(True)

# Zeigen Sie das Diagramm an
plt.show()
```
